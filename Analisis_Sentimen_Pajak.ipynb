# **TAHAP PREPOCESSING**

```python
import pandas as pd
import numpy as np
import re
import json

# Membaca data dari file CSV Anda
input_file = ('/content/tweets-data/Pajak.csv') # Ganti 'Crl_Pajak.csv' dengan nama file CSV Anda
df_busu = pd.read_csv(input_file)

# Fungsi untuk menghapus emoji menggunakan regex
def remove_emoji(text):
    if isinstance(text, str):  # Mengecek apakah 'text' adalah string
        emoji_pattern = re.compile("[\U00010000-\U0010ffff]", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)  # Menghapus emoji
    else:
        return text  # Kembalikan text jika bukan string

# Fungsi untuk menghapus username menggunakan regex
def remove_username(text):
    if isinstance(text, str):  # Mengecek apakah 'text' adalah string
        username_pattern = r'@[\w]+'  # Pola untuk mencocokkan username yang dimulai dengan '@'
        return re.sub(username_pattern, '', text)  # Menghapus username
    else:
        return text  # Kembalikan text jika bukan string

# Fungsi untuk menghapus kata ganda
def remove_duplicate_words(text):
    if isinstance(text, str):  # Mengecek apakah 'text' adalah string
        return re.sub(r'\b(\w+)\s+\1\b', r'\1', text)  # Menghapus kata ganda
    else:
        return text  # Kembalikan text jika bukan string

# Fungsi untuk menghapus spasi berlebih
def remove_extra_spaces(text):
    if isinstance(text, str):  # Mengecek apakah 'text' adalah string
        return re.sub(r'\s+', ' ', text.strip())  # Mengganti spasi berlebih dengan satu spasi
    else:
        return text  # Kembalikan text jika bukan string

# Parsing JSON dalam kolom 'full_text' jika diperlukan
def parse_json(content):
    try:
        return json.loads(content) if isinstance(content, str) else content
    except (ValueError, TypeError):
        return content  # Kembalikan jika tidak bisa diparse

# Parsing dan ekstraksi teks dari kolom 'full_text'
df_busu['full_text'] = df_busu['full_text'].apply(parse_json)

# Misalnya, jika Anda ingin mengekstrak teks dari JSON di 'full_text'
df_busu['review'] = df_busu['full_text'].apply(lambda x: x.get('content', '') if isinstance(x, dict) else x)

# Proses cleaning data
df_busu['clean_review'] = df_busu['review'].apply(remove_emoji)  # Menghapus emoji
df_busu['clean_review'] = df_busu['clean_review'].apply(remove_username)  # Menghapus username
df_busu['clean_review'] = df_busu['clean_review'].apply(remove_duplicate_words)  # Menghapus kata ganda
df_busu['clean_review'] = df_busu['clean_review'].apply(remove_extra_spaces)  # Menghapus spasi berlebih
df_busu['clean_review'] = df_busu['clean_review'].str.lower()  # Mengubah huruf menjadi kecil

# Menampilkan seluruh data yang sudah diperbaiki
pd.set_option('display.max_rows', None)  # Menampilkan seluruh baris
pd.set_option('display.max_columns', None)  # Menampilkan seluruh kolom

print(df_busu)  # Menampilkan seluruh DataFrame yang sudah diperbaiki

# Menyimpan DataFrame ke file CSV
output_file = 'Cleaned_Pajak.csv'  # File CSV hasil cleaning
df_busu.to_csv(output_file, index=False)  # Menyimpan hasil ke file CSV

print(f"Data berhasil disimpan ke CSV: {output_file}")
```

```python
import pandas as pd
import spacy

# Memuat model bahasa Inggris dari spaCy
nlp = spacy.load("en_core_web_sm")

# Membaca file CSV yang ada
df = pd.read_csv('cleaned_full_text.csv')  # Ganti dengan path file CSV Anda

# Fungsi untuk melakukan tokenisasi menggunakan spaCy
def tokenize_text_spacy(text):
    if isinstance(text, str):  # Mengecek apakah teks adalah string
        doc = nlp(text)  # Memproses teks menggunakan spaCy
        return [token.text for token in doc]  # Mengembalikan daftar token
    else:
        return []  # Jika bukan string, kembalikan list kosong

# Misalkan kolom yang ingin ditokenisasi bernama 'review' (ubah sesuai kolom Anda)
df['tokenized_review'] = df['review'].apply(tokenize_text_spacy)

# Menampilkan 5 output teratas
print(df[['review', 'tokenized_review']].head())

# Menyimpan DataFrame dengan tokenisasi ke file CSV baru
df.to_csv('tokenisasi_file.csv', index=False)  # Ganti dengan path yang Anda inginkan
```

**Normalisasi**

```python
# Contoh kamus colloquial-indonesian-lexicon
lexicon = {
    'gk': 'tidak',  # 'gk' menjadi 'tidak'
    'yg': 'yang',   # 'yg' menjadi 'yang'
    'aja': 'saja',  # 'aja' menjadi 'saja'
    'bgt': 'banget',  # 'bgt' menjadi 'banget'
    'ga': 'tidak',   # 'ga' menjadi 'tidak'
    'msh': 'masih',  # 'msh' menjadi 'masih'
    'lo': 'kamu',    # 'lo' menjadi 'kamu'
    'gue': 'saya',   # 'gue' menjadi 'saya'
    'dapet': 'dapat' # 'dapet' menjadi 'dapat'
}
```

```python
import re

# Fungsi untuk normalisasi kata non-formal menjadi formal
def normalize_text(text, lexicon):
    # Tokenisasi kalimat menjadi kata-kata
    words = text.split()

    # Normalisasi setiap kata
    normalized_words = [lexicon.get(word.lower(), word) for word in words]

    # Gabungkan kembali kata-kata menjadi kalimat
    return ' '.join(normalized_words)
```

```python
import pandas as pd

# Misalkan data Anda ada di file CSV
df = pd.read_csv('tokenisasi_file.csv')  # Ganti dengan path file Anda

# Terapkan normalisasi ke kolom 'review'
df['normalized_review'] = df['review'].apply(lambda x: normalize_text(x, lexicon))

# Menampilkan beberapa baris pertama untuk melihat hasilnya
print(df[['review', 'normalized_review']].head())

# Menyimpan DataFrame yang sudah dinormalisasi ke file CSV baru
df.to_csv('normalized_reviews.csv', index=False)
```

**Konvensi Negas**


```python
# Daftar kata-kata negasi yang umum dalam bahasa Indonesia
negation_words = ['tidak', 'bukan', 'ga', 'belum', 'jangan', 'gak', 'nggak', 'kurang', 'sama sekali']
```

```python
import re

# Fungsi untuk menandai kata-kata yang terpengaruh negasi
def convert_negation(text):
    words = text.split()  # Memecah kalimat menjadi kata-kata
    result = []  # Daftar untuk menampung hasil konversi

    # Variabel untuk menentukan apakah kata berikutnya terpengaruh negasi
    is_negated = False

    for i, word in enumerate(words):
        # Cek apakah kata adalah kata negasi
        if word.lower() in negation_words:
            is_negated = True  # Set negasi aktif setelah kata negasi
            result.append(word)  # Menambahkan kata negasi ke hasil
        else:
            if is_negated:
                # Mengubah kata setelah negasi menjadi negatif (contoh: "baik" menjadi "tidak baik")
                negated_word = "tidak " + word
                result.append(negated_word)  # Menambahkan kata yang sudah ternegasi
                is_negated = False  # Reset flag negasi setelah kata ternegasi
            else:
                result.append(word)  # Menambahkan kata yang tidak terpengaruh negasi

    return ' '.join(result)  # Gabungkan kembali menjadi kalimat
```

```python
import pandas as pd

# Misalkan data Anda ada di file CSV
df = pd.read_csv('normalized_reviews.csv')  # Ganti dengan path file Anda

# Terapkan konversi negasi ke kolom 'clean_review'
df['negated_review'] = df['clean_review'].apply(convert_negation)

# Menampilkan beberapa baris pertama untuk melihat hasilnya
print(df[['clean_review', 'negated_review']].head())

# Menyimpan DataFrame dengan konversi negasi ke file CSV baru
df.to_csv('negated_reviews.csv', index=False)
```

```python
from nltk.corpus import stopwords

# Menyediakan daftar stopwords dari NLTK
stop_words = set(stopwords.words('indonesian'))  # Menggunakan bahasa Indonesia

# Fungsi untuk menghapus stopwords dari teks
def remove_stopwords(text):
    # Memecah kalimat menjadi kata-kata
    words = text.split()

    # Menghapus stopwords dengan memfilter kata-kata yang ada dalam stop_words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Gabungkan kembali kata-kata yang sudah difilter menjadi kalimat
    return ' '.join(filtered_words)
```

```python
import pandas as pd

# Misalkan data Anda ada di file CSV
df = pd.read_csv('negated_reviews.csv')  # Ganti dengan path file Anda

# Terapkan stopword removal ke kolom 'clean_review'
df['review_no_stopwords'] = df['clean_review'].apply(remove_stopwords)

# Menampilkan beberapa baris pertama untuk melihat hasilnya
print(df[['clean_review', 'review_no_stopwords']].head())

# Menyimpan DataFrame dengan stopword removal ke file CSV baru
df.to_csv('reviews_no_stopwords.csv', index=False)
```

Pembersihan noise setelah melakukan convert negation, untuk menghapus karakter _ dengan modul re.

```python
import pandas as pd
import re

# Fungsi untuk menghapus karakter tertentu (misalnya '_') dalam teks
def remove_noise(text):
    if isinstance(text, str):  # Mengecek apakah 'text' adalah string
        # Menghapus karakter '_'
        text = re.sub(r'_', '', text)  # Hapus semua karakter '_'
        # Tambahkan aturan pembersihan lain jika diperlukan (misal menghapus karakter lain)
        return text
    else:
        return text  # Kembalikan text jika bukan string

# Membaca data dari CSV (misalnya 'reviews.csv')
df = pd.read_csv('reviews_with_sentences.csv')  # Ganti dengan path ke file CSV Anda

# Misalkan kolom yang perlu dibersihkan adalah 'clean_review'
# Terapkan fungsi 'remove_noise' ke seluruh kolom 'clean_review'
df['clean_review'] = df['clean_review'].apply(remove_noise)

# Menyimpan DataFrame yang sudah dibersihkan ke file CSV baru
df.to_csv('/content/reviews_with_sentences.csv', index=False)

# Menampilkan beberapa baris pertama dari DataFrame yang sudah dibersihkan
print(df[['clean_review']].head())
```

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi stemmer menggunakan Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada setiap teks
def stemming_teks(teks):
    return stemmer.stem(teks)

# Membaca file CSV yang berisi data ulasan
df = pd.read_csv('data_clean_sentimen.csv')  # Gantilah dengan path file CSV Anda

# Pastikan kolom ulasan bernama 'full_text'
if 'full_text' not in df.columns:
    raise ValueError("Pastikan file CSV memiliki kolom bernama 'full_text'.")

# Melakukan stemming pada kolom full_text
df['full_text_stemmed'] = df['full_text'].apply(stemming_teks)

# Menghitung jumlah kata yang telah di-stem untuk setiap dokumen
df['jumlah_kata_stemmed'] = df['full_text_stemmed'].apply(lambda x: len(x.split()))

# Menampilkan data hasil perhitungan jumlah kata setelah stemming
print("Data setelah perhitungan jumlah kata setelah stemming:")
print(df[['full_text', 'full_text_stemmed', 'jumlah_kata_stemmed']])
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Membaca data dari file CSV (gantilah dengan path file CSV Anda)
df = pd.read_csv('data_clean_sentimen.csv')  # Ganti dengan path file Anda

# Misalkan kolom yang berisi ulasan adalah 'content'
teks = df['full_text']

# Menghitung TF-IDF menggunakan TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(teks)

# Mendapatkan kata-kata dan id dokumen
words = vectorizer.get_feature_names_out()
docs = tfidf_matrix.toarray()

# Menyusun data untuk DataFrame
data = []

for doc_index, doc in enumerate(docs):
    for word_index, score in enumerate(doc):
        if score > 0:  # Hanya masukkan kata yang memiliki nilai TF-IDF lebih dari 0
            data.append({
                'jumlah_dokumen': doc_index + 1,  # Menyebutkan urutan dokumen (1-based index)
                'letak_kata': words[word_index],
                'tf_idf': score
            })

# Membuat DataFrame
df_tfidf = pd.DataFrame(data)

# Menampilkan DataFrame
print("Hasil TF-IDF:")
print(df_tfidf)
```

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi stemmer menggunakan Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada setiap teks
def stemming_teks(teks):
    return stemmer.stem(teks)

# Membaca file CSV yang berisi data ulasan
df = pd.read_csv('data_clean_sentimen.csv')  # Gantilah dengan path file CSV Anda

# Pastikan kolom ulasan bernama 'full_text' atau 'clean_review'
if 'full_text' not in df.columns and 'clean_review' not in df.columns:
    raise ValueError("Pastikan file CSV memiliki kolom bernama 'full_text' atau 'clean_review'.")

# Gunakan kolom yang tersedia, 'full_text' atau 'clean_review'
kolom_utama = 'full_text' if 'full_text' in df.columns else 'clean_review'

# Melakukan stemming pada kolom utama
df['full_text_stemmed'] = df[kolom_utama].apply(stemming_teks)

# Menghitung jumlah kata yang telah di-stem untuk setiap dokumen
df['jumlah_kata_stemmed'] = df['full_text_stemmed'].apply(lambda x: len(x.split()))

# Menampilkan data hasil perhitungan jumlah kata setelah stemming
print("Data setelah perhitungan jumlah kata setelah stemming:")
print(df[[kolom_utama, 'full_text_stemmed', 'jumlah_kata_stemmed']])
```
